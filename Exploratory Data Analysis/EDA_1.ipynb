```python
import seaborn as sns
def correlation_heatmap(df):
    _,ax=plt.subplots(figsize=(20,20))
    colormap=sns.color_palette("BuGn_r")
    sns.heatmap(df.corr(),annot=True,cmap=colormap)
    
    
correlation_heatmap(df)
```
